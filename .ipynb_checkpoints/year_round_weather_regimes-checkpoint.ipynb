## Computing seven NAE weather regimes

The first step involves computing the seven year-round Atlantic-European weather regimes first defined in [Grams et al., 2017](https://www.nature.com/articles/nclimate3338) and later in [Büeler et al., 2021](https://rmets.onlinelibrary.wiley.com/doi/full/10.1002/qj.4178).  


This can be summarized in the following steps:
1. Compute six-hourly 500-hPa geopotential height anomalies with respect to 91-day running mean calendar date climatologies (i.e., +/- 45 days centered around each 6-hr time step)
2. Filter anomalies using five-day low pass filter
3. Seasonally normalize (divide low-pass-filtered geopotential height anomaly at each grid point by a calendar-day-dependent scalar that quantifies the climatological variability of geopotential height anomalies at the corresponding calendar day. This scalar is computed as the spatial average (over all grid points in the investigated domain) of the temporal 31-day running standard deviation over all anomalies between 1979 and 2018.)
4. Apply EOF analysis to filtered and seasonally normalized anomalies based on the North-Atlantic-European domain (80W to 40E, 30 to 90N)
5. Apply k-means clustering in the phase space spanned by the first seven EOFs (70% variance), yielding seven cluster means representing the seven weather regimes

**These regimes are called**

a) Atlantic Trough (AT)  

b) Zonal Regime (ZO)

c) Scandinavian Trough (ScTr)

d) Atlantic Ridge (AR)

e) European Blocking (EuBl)

f) Scandinavian Blocking (ScBL)

g) Greenland Blocking (GL)

h) No Regime

### Load and preprocess ERA data

```python
path_pre = './data/pre/156/'
path_pos = './data/pos/156/'

preprocess = False
```

```python
def regrid(data, ddeg_out=1., method='bilinear', reuse_weights=False, filename='grid.nc'):
    """Convert to 1 degree res"""

    ## Create output grid
    grid_out = xr.Dataset(
        {
            'lon': (['lon'], np.arange(-80, 40 + ddeg_out, ddeg_out)),
            'lat': (['lat'], np.arange(20, 90, ddeg_out)),
        }
    )

    # Create regridder
    regridder = xe.Regridder(
        data, grid_out, method, periodic=True, reuse_weights=reuse_weights,
        filename=filename
        )

    return regridder(data, keep_attrs=True).astype('float32')
```

```python
if preprocess:
    years = np.arange(1979, 2020)

    for year in years:
        # Retrieve file paths
        files = glob.glob(path_pre + '{}/*500hPa*.grb2'.format(year))
        files.sort()

        # Load data
        z_actual = xr.open_mfdataset(files, engine='cfgrib')
        z_actual = z_actual['z']

        # Convert to 6-hourly
        datetime_indices = pd.date_range(z_actual.time.values[0], z_actual.time.values[-1], freq='6H')
        z_actual = z_actual.sel(time=datetime_indices)

        # Load into memory 
        z_actual = z_actual.load()

        # Regrid to 1 deg res for NAE region
        z_actual = regrid(z_actual)

        # Save to pos folder
        z_actual.to_netcdf('{}{}/era5_ecmwf_ra_dt_gph500hPa_1deg_nae_{}.nc'.format(path_pos, year, year))
```

### Compute six-hourly 500-hPa geopotential height anomalies with respect to 91-day running mean calendar date climatologies (i.e., +/- 45 days centered around each 6-hr time step)

#### Compute climatology

```python
z_actual = z_actual.load()
```

```python
z_clim = z_actual.rolling(time=91, center=True).mean()
z_anom = (z_actual - z_clim)

z_anom = z_anom.dropna(dim='time')
```

```python
# Convert to 4D with level
z_anom = xr.DataArray(np.expand_dims(z_anom.values, 1), dims=['time', 'level', 'lat', 'lon'],
            coords={'lat': z_anom.lat,
                    'lon': z_anom.lon,
                    'time': z_anom.time,
                    'level': [500.]})
```

### Seasonally normalize

Divide low-pass-filtered geopotential height anomaly at each grid point by a calendar-day-dependent scalar that quantifies the climatological variability of geopotential height anomalies at the corresponding calendar day. This scalar is computed as the spatial average (over all grid points in the investigated domain) of the temporal 31-day running standard deviation over all anomalies between 1979 and 2018.

```python
z_anom = z_anom / z_anom.rolling(time=31, center=False).std()
z_anom = z_anom.dropna(dim='time')
```

### Apply EOF analysis to filtered and seasonally normalized anomalies based on the North-Atlantic-European domain

```python
# Compute coslat correction weights
coslat = np.cos(np.deg2rad(z_anom.coords['lat'].values)).clip(0., 1.)
wgts = np.sqrt(coslat)[..., np.newaxis]
use_scaling = 0
n_eofs = 7
n_eofs_for_kmeans = 7
n_clusters = 7
```

```python
# EOF solver
solver = Eof(z_anom, weights=wgts) # wgts
eofs = solver.eofs(neofs=n_eofs)
eofs = eofs * (-1)

eofs_corr = solver.eofsAsCorrelation(neofs=n_eofs)
eofs_corr = eofs_corr * (-1)

explained_var = solver.varianceFraction()
```

```python
# Save principal components
pcs = solver.projectField(z_anom, neofs=n_eofs, eofscaling=use_scaling)
# pcs = solver.pcs(npcs=20, pcscaling=use_scaling)

# changing sign of eofs and pcs depending on solution
pcs = pcs * (-1)

# Convert to pd.dataframe
df_pcs = pcs.to_dataframe().reset_index()
df_pcs = df_pcs.pivot(index='time', columns='mode',values='pseudo_pcs')
df_pcs = df_pcs.add_prefix('EOF')
df_pcs.index = pd.to_datetime(df_pcs.index)
```

### Apply k-means clustering in the phase space spanned by the first seven EOFs (70% variance), yielding seven cluster means representing the seven weather regimes


```python
regimes = ['GL', 'AT', 'AR?',  'ScTr', 'EuBL', 'ScBL', 'ZO' ]
```

```python
def cluster_index_to_regime(x, regimes = ['GL', 'AT', 'AR?',  'ScTr', 'EuBL', 'ScBL', 'ZO' ]):       
    return regimes[x]
```

```python
# Fit k-means
X_train = df_pcs.iloc[:, 0:n_eofs_for_kmeans].values
km = KMeans(n_clusters=n_clusters, n_init=500, verbose=0, tol=0.000001, max_iter=500, random_state=3)
km.fit(X_train)
```

```python
# Compute clusters on reanalysis
clus = pd.DataFrame(data=km.labels_, columns=['cluster_id'], index=df_pcs.index).reset_index()
clus.rename(columns={'time': 'date'}, inplace=True)
clus['cluster_regime'] = clus['cluster_id'].apply(cluster_index_to_regime)
clus.set_index('date', inplace=True)
```

```python
# Renormalize and save clusters into xarray

cluster_centers = pd.DataFrame(km.cluster_centers_, columns=[f'eof{i}' for i in np.arange(1,n_eofs_for_kmeans+1)])
cluster_centers.index = regimes
# Creating a Data array containing the cluster_centers in the PCA space
cluster_center_array = xr.DataArray(cluster_centers.values, 
                                    coords=[np.arange(0, n_clusters), np.arange(0, n_eofs_for_kmeans) ], 
                                    dims=['centroids', 'mode'])

# Computing the weighted cluster centers in the original lat-lon space
eigenvectors20 = eofs.loc[0:n_eofs_for_kmeans, ...]
recons_cluster_centers_weighted = cluster_center_array.dot(eigenvectors20)

lons = recons_cluster_centers_weighted.coords['lon'].values 
lats = recons_cluster_centers_weighted.coords['lat'].values

coslat = np.cos(np.deg2rad(lats)).clip(0., 1.)
wgts = np.sqrt(coslat)

weights_array = xr.DataArray(wgts, coords=[lats], dims=['lat'])

# Dividing by the weights and by the globally averaged acceleration of gravity at sea level (g0 = 9.80665 m s−2)
# in order to get the results in meters (original fields are Z*g0 [m2s-2])
g0 = 9.80665
recons_cluster_centers_unweighted = recons_cluster_centers_weighted / weights_array / g0

# Set index correctly for cluster
clus.index = pd.to_datetime(z_anom.time.values)
# clus.index = pd.to_datetime(z_allyear.indexes['time'].to_datetimeindex())
```

#### Plot clusters

```python
for i in np.arange(0, n_clusters):

    plt.figure(figsize=(15, 8))
    ax = plt.axes(
        projection=ccrs.Orthographic(central_longitude=0, central_latitude=45))
    
    vals = recons_cluster_centers_unweighted[i, :, :].squeeze().values

    fill = ax.contourf(
        lons,
        lats,
        recons_cluster_centers_unweighted[i, :, :].squeeze(),
        levels=18,
        norm=TwoSlopeNorm(
            vmin=np.nanmin(vals[~np.isinf(vals)]),
            vcenter=0,
            vmax=np.nanmax(vals[~np.isinf(vals)])),
        cmap=plt.cm.RdBu_r,
        transform=ccrs.PlateCarree())

    cb = plt.colorbar(fill,
                      orientation='vertical',
                      label='500hPa gph anomaly [m]')
    title = f'cluster center: {cluster_index_to_regime(i)}'

    ax.coastlines(resolution='50m')
    ax.gridlines()
    
    plt.title(title, fontsize=20)
    plt.show()
```
